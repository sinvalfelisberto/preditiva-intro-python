## Aplicação 1

Problema: Você deseja verificar o comportamento de distribuição uniforme no sorteio de números da mega-sena. Para isso faça uma simulação de 10 milhão de sorteios e depois verifique a frequência com que cada número foi sorteado.

```python
import random
import matplotlib.pyplot as plt
```

```python
for i in range(1, 61):
    print(i)
```

```python
sorteio = random.sample(range(1, 61), 6)
print(sorteio)
```

```python
qtd_vezes_sorteio = []
for i in range(60):
    qtd_vezes_sorteio.append(0)

qtd_vezes_sorteio
```

```python
for _ in range(10**7):
    sorteio = random.sample(range(1, 61), 6)
    for s in sorteio:
        qtd_vezes_sorteio[s-1] = qtd_vezes_sorteio[s-1] + 1
```

```python
qtd_vezes_sorteio
```

```python
plt.bar(range(1,61), qtd_vezes_sorteio)
```
