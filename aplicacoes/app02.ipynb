````python
import pandas as pd

file = pd.read_excel('../files/bank_marketing.xlsx', skiprows=1)
file = file[['Cliente_Comprou_o_Titulo?',	'Idade',	'Profissão',	'Estado_Civil',	'Formação',	'Cliente_Devedor?',	'Saldo_Conta_Corrente',	'Tem_Hipoteca?',	'Tem_Emprestimo?',	'Qte_de_Ligações_Feitas']]
dados = file.to_dict('records')
````

````python
dados[:2]
````

````python
qtd_clientes = len(dados)

qtd_clientes
````

````python
clientes_compraram = 0
for cliente in dados:
    if cliente['Cliente_Comprou_o_Titulo?'] == 'Sim':
        clientes_compraram = clientes_compraram + 1

clientes_compraram
````

````python
perc = clientes_compraram / qtd_clientes * 100

perc
````
