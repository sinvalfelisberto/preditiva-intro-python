## Classes

- Quantidade de Dados
- Calcular a média da amostra
- Calcular a variânbcia amostral
- Calcular uma estimativa para a média populacional


```python
class Amostras:
    def __init__(self, lista_dados):
        self.dados = lista_dados

    def media_amostral(self):
        '''
        Retorna a média amostral de uma amostra.

        
        Returns:
            float: média amostral da lista de dados da amostra.

        Example:
            >>> amostra = Amostras([6, 13, 2, 13, 15, 4, 16, 6, 18, 22, 22, 22, 22, 22, 22,])
            >>> print(amostra.media_amostral)
            10.6
        '''
        qtd_dados = len(self.dados)
        soma_dados = sum(self.dados)
        return soma_dados / qtd_dados
```

```python
class Pessoa:
    def __init__(self, nome, data_nascimento, nome_mae):
        self.nome = nome
        self.data_nascimento = data_nascimento
        self.nome_mae = nome_mae

    def __str__(self):
        return f'nome: {self.nome}\ndata_nascimento: {self.data_nascimento.replace('-','/')}\nnome mãe: {self.nome_mae}'
```
