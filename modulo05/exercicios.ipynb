**Problema**: Suponha que você receba o nome de uma cidade no formato "Cidade - UF", crie uma função para retornar o nome da cidade em uma tupla com primeiro elemento sendo o estado e o segundo elemento sendo a cidade


```python
def retorna_uf_cidade(cidade: str):
    '''
    recebe um texto de CIDADE-UF e retorna uma tupla com UF e Município.

    Args:
        cidade (str): o nome da cidade e seu estado.

    Returns:
        tuple: tupla no formato (uf, cidade).

    Example:
        >>> cidade = 'Brasilia-DF'
        >>> retorna_uf_cidade(cidade)
        ('DF', 'Brasilia')
    '''
    cidade.lower()
    return (cidade[-2:].upper(),cidade[:-3].capitalize())
```

# Exercicio 2
Crie uma função para calcular o rendimento de um investimento a juros compostos após um determinado período de meses.

$ Retorno = Investimento Inicial * (1 + Taxa Juros)^{meses_aplicacao} $

```python
def retorno_investimento(investimento_inicial: float,taxa_juros:float,  meses: int):
    """
    Calcula o valor futuro de um investimento com juros compostos.

    O cálculo é feito utilizando a fórmula:

        valor_final = investimento_inicial * (1 + taxa_juros) ** meses

    Parameters
    ----------
    investimento_inicial : float
        Valor inicial investido.
    taxa_juros : float
        Taxa de juros mensal (em formato decimal).
        Exemplo: 2.5 representa 2,5% ao mês.
    meses : int
        Quantidade de meses em que o investimento ficará aplicado.

    Returns
    -------
    float
        Valor final do investimento após o período informado.

    Example
    -------
    >>> retorno_investimento(1000, 2, 12)
    1268.2417945625455
    """
    taxa_juros = taxa_juros / 100.0
    return investimento_inicial * (1 + taxa_juros) ** meses
```

```python
print(retorno_investimento(1000, 2, 12))
```
