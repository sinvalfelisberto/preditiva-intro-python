Substitua ```{Data}``` pela data de interesse no formato mês-dia-ano

````python
data = data = '07-01-2024'

url = f'https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/CotacaoDolarDia(dataCotacao=@dataCotacao)?@dataCotacao=%27{data}%27&$format=json'
````

````python
import requests
````

````python
response = requests.get(url)
dados = response.json()


cotacao_dolar = dados['value'][0]['cotacaoCompra']

cotacao_dolar
````
