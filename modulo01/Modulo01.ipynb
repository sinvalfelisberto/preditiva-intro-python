#### Aplicações:
Um ativo financeiro rende 5% ao mês. Investindo R$ 1.000,00 neste ativo no início de janeiro, qual o valor no final de dezembro?

```python
def jurosCompostos(capital, juros, tempo):
    return capital * (juros) ** tempo

cap = 1000
j = 1.05
temp = 12

total = jurosCompostos(cap, j, temp)

print(total)
```

#### Porcentagem
Do total de 150 alunos, 30% gostam de matemática. Quantos alunos representam esse percentual?


```python
total_alunos = 150
total_alunos_matematica = (total_alunos * 30) / 100.0

print(f"O total de alunos que gostam de matemática é {total_alunos_matematica}." )
```

#### Lucro
Se o lucro de uma empresa pode ser escrito usando a função a seguir, qual a quantidade de unidades vendidas necessparia para o lucro de R$ 10.000,00?

$ Lucro(unidades) = 400 * unidades - 30 $


```python
# 10000 = 400 * unidades - 30

def retornaUnidades(lucro):
    return (lucro + 30) // 400 + 1

lucro = 10000
unidades_vendidas = retornaUnidades(lucro)

print(f"O total de peças vendidas é {unidades_vendidas}")
```
