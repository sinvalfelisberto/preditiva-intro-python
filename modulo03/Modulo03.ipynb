**Problema**:

Você recebeu uma estrutura de dados com informações a respeito de cada cliente e sabendo que seu time de análise de dados criou uma segmentação para os clientes com a seguinte regra:

* Key Account: Faturamento acima de 1.000.000
* Prioritário: Faturamento acima de 500.000 e até 1.000.000
* Cauda Longa: Faturamento até 500.000

Você deve atualizar sua estrutura de dados de tal maneira que a segmentação dos clientes esteja presente para cada um deles.


```python
fat_clientes = {'Farma SA': 1500000,
 'Padaria Pao Bom': 250000,
 'Lojas Veste Mais': 800000
 }
```

```python
class Cliente:
    def __init__(self, nome, faturamento, classificacao):
        self.nome = nome
        self.faturamente = faturamento
        self.classificacao = classificacao


clientes_classificados = []
```

```python
for info_cliente in fat_clientes.items():
    if info_cliente[1] <= 500000:
        clientes_classificados.append(Cliente(info_cliente[0], info_cliente[1], 'Cauda Longa'))
    elif info_cliente[1] <= 1000000:
        clientes_classificados.append(Cliente(info_cliente[0], info_cliente[1], 'Prioritário'))
    else:
        clientes_classificados.append(Cliente(info_cliente[0], info_cliente[1], 'Key Account'))
        
```

```python
for cliente in clientes_classificados:
    print(cliente.nome , cliente.faturamente, cliente.classificacao)
```
