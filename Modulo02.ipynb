#### Velocidade e Memória

##### Byte
O byte é composto por 8 bits, sendo o bit uma abreviação de binary digit (0 or 1).

Preparação

Para verificar o quanto de memória é usado para armazenar cada variável em Python, vamos usar a função sys.getsizeof()

#

````python
import sys

teste = True

print(f"O tamanho da variável teste é {sys.getsizeof(teste)} bytes")

print(f"O tamanho do número 2**1000 {sys.getsizeof(2**1000)} bytes")

print(f"O tamanho do texto Sinval Felisberto {sys.getsizeof('Sinval Felisberto')} bytes")
````

### Listas, Tuplas, Conjuntos e Dicionários

## Tupla

````python
nome = "Sinval"
data_nascimento = '01/01/1982'
cpf = "012.345.678-90"
nome_da_mae = 'zulmira'
idade = 43

# tuplas são imutáveis --> não há como atualizar o valor de uma tupla
pessoa = nome, data_nascimento, cpf, nome_da_mae, idade

print(pessoa)
print(type(pessoa))
print('nome da mae: ' +pessoa[3])
print(pessoa[1])

````

## Conjuntos

Guarda os dados sem duplicação. Não são ordenados e não mutáveis.

````python
palavras_mencionadas = {
    'Simplesmente', 'o', 'melhor', 'curso', 'de', 'análise', 'e', 'ciência', 'de', 'dados', 'do', 'mercado',
    'A', 'Preditiva', 'foi', 'a', 'melhor', 'escolha', 'para', 'os', 'cursos', 'de', 'dados',
    'Gratidão', 'por', 'ter', 'encontrado', 'a', 'Preditiva', 'na', 'minha', 'jornada', 'de', 'alfabetização', 'em', 'dados'
}

print(palavras_mencionadas)

print(type(palavras_mencionadas))
````

````python
print('dados' in palavras_mencionadas)
print('Excel' in palavras_mencionadas)

palavras_mencionadas.add('Excel')

print('Excel' in palavras_mencionadas)

print(palavras_mencionadas)
````
