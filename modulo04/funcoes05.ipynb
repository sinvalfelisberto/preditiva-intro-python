## Funções 

### Lambda

Pequenas **funções anônimas** que podem ser criadas com a palavra-chave **lambda**

```python
    sqrt = lambda x: x ** (1/2)
```

````python
sqrt = lambda x: x ** (1/2)
````

````python
sqrt(100)
````

````python
f = lambda x: 'Presencial' if x == 0 else 'Remoto' if x == 100 else 'Híbrido'
````

````python
lista_classificacao = list(map(f, [1, 100, 0, 50, 20, 50, 100, 0]))

print(lista_classificacao)
````
