```python
file = open('../files/Dataset salary 2024.csv')

lines = []


for line in file:
    lines.append(line)


lines[:20]
```

```python
fields = lines[0].split(',')
lines = lines[1:]

fields
```

```python
fields[-1] = fields[-1][:-1]
```

```python
fields
```

```python
dados = {}

for campo in fields:
    dados[campo] = []
```

```python
dados
```

```python
for line in lines:
    infos = line.split(',')

    if len(infos) != len(fields):
        pass
    else:
        for i in range(len(fields)):
            if i == len(fields) -1:
                infos[i] = infos[i][:-1]
            dados[fields[i]].append(infos[i])

dados.keys()

```
